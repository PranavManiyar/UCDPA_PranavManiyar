# Machine Learning





```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
#Reading the base data file
df_student=pd.read_csv('student-mat.csv',sep=";")
```

```python
df_student.head()
```

```python
df_student.info()
```

```python
#keeping the only varibales that we would use
Student= df_student[["G1", "G2", "G3", "studytime", "failures", "absences",'health' ]]
```

```python
Student.head(10)
```

```python
predict = "G3"

X = np.array(Student.drop([predict], 1)) # Features
y = np.array(Student[predict]) # Labels
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

```python
linear = linear_model.LinearRegression()
```

```python
linear.fit(x_train, y_train)
acc = linear.score(x_test, y_test) # acc stands for accuracy 
```

```python
print(acc)
```

```python
print('Coefficient: \n', linear.coef_) # These are each slope value
print('Intercept: \n', linear.intercept_) # This is the intercept
```

```python
predictions = linear.predict(x_test) # Gets a list of all predictions

for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])
```
